```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import re
```

```python
df = pd.read_csv("../Inputs/phoneticDictionary.csv")
```

```python
syl_df = df[['word','syl']]
```

```python
def count_vowels(word):
    vowels = 0
    for letter in word:
        if letter in ['a','e','i','o','u']:
            vowels += 1      

    return vowels

def count_vowel_groups(word):
    word = word.lower()
    groups = len(re.findall(r'[aeiouy]+', word))
    if word.endswith('e'):
        groups -= 1
    return max(groups, 1)
```

```python
syl_df = syl_df.dropna()
syl_df['length'] = syl_df['word'].str.len()
syl_df['vowel_count'] = syl_df['word'].apply(count_vowels)
syl_df['non_vowel_count'] = syl_df['length'] - syl_df['vowel_count']
syl_df['vowel_groups'] = syl_df['word'].apply(count_vowel_groups)

syl_df.head()
```

```python
X = syl_df[['length','vowel_count','vowel_groups']]
y = syl_df[['syl']]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

type(X_test)
```

```python
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
scores = cross_val_score(reg, X, y, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
