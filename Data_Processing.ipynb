```python
bi = pd.read_csv("Data/BillieEilish.csv")
syl = pd.read_csv("Data/phoneticDictionary.csv")
```

```python

syl_cols_to_keep = ["word", "syl"]
syl = syl[syl_cols_to_keep]
print(syl[syl["word"] == 'white'])
```

```python
bad_guy_lyrics = bi["Lyric"][2].split()
everything_i_wanted_lyrics = bi["Lyric"][1].split()
print(bad_guy_lyrics)
```

```python
bad_guy_lyric_syls = {}
everything_lyric_syls = {}
for item in bad_guy_lyrics:
    
    vals1 = syl.loc[syl['word'] == item, 'syl']
    #print(f"{item}: {vals1.iloc[0] if not vals1.empty else None}")
    bad_guy_lyric_syls[item] = vals1.iloc[0] if not vals1.empty else None

for item in everything_i_wanted_lyrics:
    
    vals2 = syl.loc[syl['word'] == item, 'syl']
    print(f"{item}: {vals2.iloc[0] if not vals2.empty else None}")
    everything_lyric_syls[item] = vals2.iloc[0] if not vals2.empty else None

```

```python
#fig, ax = plt.subplots()
#ax.plot(bad_guy_lyric_syls.values())
plt.plot(bad_guy_lyric_syls.values(),color = 'red')
plt.plot(everything_lyric_syls.values(),color = 'blue')
```
