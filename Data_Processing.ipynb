```python
bi = pd.read_csv("Data/BillieEilish.csv")
syl = pd.read_csv("Data/phoneticDictionary.csv")
```

```python

syl_cols_to_keep = ["word", "syl"]
syl = syl[syl_cols_to_keep]
print(syl[syl["word"] == 'white'])
```

```python
bad_guy_lyrics = bi["Lyric"][2]
bad_guy_lyrics_split = bad_guy_lyrics.split()
print(bad_guy_lyrics_split)
```

```python
lyric_syls = {}
for item in bad_guy_lyrics_split:
    
    vals = syl.loc[syl['word'] == item, 'syl']
    print(f"{item}: {vals.iloc[0] if not vals.empty else None}")
    lyric_syls[item] = vals.iloc[0] if not vals.empty else None

```

```python
for key,value in lyric_syls.items():
    print(key,value)
```
